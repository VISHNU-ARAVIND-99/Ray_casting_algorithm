```python
def point_inside_a_polygon_or_not(point, polygon):
    x, y = point
    n = len(polygon)
    count = 0
    for i in range(n):
        x1, y1 = polygon[i]
        x2, y2 = polygon[(i + 1) % n]
        if (y1 > y) != (y2 > y) and (x < x1 + (y - y1) / (y2 - y1) * (x2 - x1)):
            count += 1
    if count % 2 == 1:
        return "Inside"
    else:
        return "Outside"

# Example usage
point = (5, 6)
polygon = [(0, 0), (7, 0), (7, 7)]
print(point_inside_a_polygon_or_not(point, polygon))
```
